 ### a. Missing Value

```python
import pandas as pd
import numpy as np
```

```python
dataMV=pd.read_csv("bottle DATAFIX.csv",sep=",")
dataMV
```

```python
# statistika deskriptif
dataMV.describe()
```

```python
# Missing Value
np.sum(dataMV.isnull())
```

```python
dataMV['R_POTEMP']=dataMV['R_POTEMP'].fillna((dataMV['R_POTEMP'].mean()))
dataMV['R_TEMP']=dataMV['R_TEMP'].fillna((dataMV['R_TEMP'].mean()))
dataMV['R_SALINITY']=dataMV['R_SALINITY'].fillna((dataMV['R_SALINITY'].mean()))
dataMV['R_SIGMA']=dataMV['R_SIGMA'].fillna((dataMV['R_SIGMA'].mean()))
dataMV['R_SVA']=dataMV['R_SVA'].fillna((dataMV['R_SVA'].mean()))
dataMV['R_DYNHT']=dataMV['R_DYNHT'].fillna((dataMV['R_DYNHT'].mean()))
dataMV['R_O2']=dataMV['R_O2'].fillna((dataMV['R_O2'].mean()))
dataMV['R_O2Sat']=dataMV['R_O2Sat'].fillna((dataMV['R_O2Sat'].mean()))
np.sum(dataMV.isnull())
```

```python
dataMV
```

```python
import csv
np.savetxt('dataout_a.csv',dataMV, delimiter=',', fmt='%f'
 ,header="R_POTEMP,R_TEMP,R_DEPTH,R_SALINITY,R_SIGMA,R_SVA,R_DYNHT,R_PRES,R_O2,R_O2Sat")
```

### b. Missing Value dan Outlier

```python
# mengatasi outlier univariat
from scipy import stats
z=np.abs(stats.zscore(dataMV._get_numeric_data()))
print(z)
dataMV=dataMV[(z<3).all(axis=1)]
print(dataMV.shape)
```

```python
np.savetxt('dataout_b.csv',dataMV, delimiter=',', fmt='%f'
 ,header="R_POTEMP,R_TEMP,R_DEPTH,R_SALINITY,R_SIGMA,R_SVA,R_DYNHT,R_PRES,R_O2,R_O2Sat")
```

### c. Missing Value dan Transformasi

```python
import pandas as pd
import numpy as np
data_c=pd.read_csv("dataout_a.csv",sep=",")
data_c
```

```python
x=data_c.iloc[0:5000,1:10] #iloc buat ambil data ke-
x.tail()
```

```python
from sklearn.preprocessing import normalize
xnorm=normalize(x,norm='l2') #default norm l2, other are l1 and max
xnorm=pd.DataFrame(xnorm)
xnorm.tail()
```

```python
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x),axis=0)
xsoft=softmax(xnorm)
xsoft=pd.DataFrame(xsoft)
print(xsoft)
```

```python
np.savetxt('datatrans1.csv',xsoft, delimiter=',', fmt='%f'
 ,header="R_TEMP,R_DEPTH,R_SALINITY,R_SIGMA,R_SVA,R_DYNHT,R_PRES,R_O2,R_O2Sat")
```
