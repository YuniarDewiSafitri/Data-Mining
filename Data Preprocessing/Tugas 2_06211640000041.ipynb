```python
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv("deepsolar_t.csv",sep=",")
data1
```

```python
# statistika deskriptif
data1.describe()
```

```python
#missing value
np.sum(data1.isnull())
```

```python
#imputasi untuk missing value
data1['average_household_income']=data1['average_household_income'].fillna((data1['average_household_income'].mean()))
data1['gini_index']=data1['gini_index'].fillna((data1['gini_index'].mean()))
data1['land_area']=data1['land_area'].fillna((data1['land_area'].mean()))
data1['per_capita_income']=data1['per_capita_income'].fillna((data1['per_capita_income'].mean()))
data1['population_density']=data1['population_density'].fillna((data1['population_density'].mean()))
data1['travel_time_less_than_10_rate']=data1['travel_time_less_than_10_rate'].fillna((data1['travel_time_60_89_rate'].mean()))
data1['travel_time_10_19_rate']=data1['travel_time_10_19_rate'].fillna((data1['travel_time_10_19_rate'].mean()))
data1['travel_time_20_29_rate']=data1['travel_time_20_29_rate'].fillna((data1['travel_time_20_29_rate'].mean()))
data1['travel_time_30_39_rate']=data1['travel_time_30_39_rate'].fillna((data1['travel_time_30_39_rate'].mean()))
data1['travel_time_40_59_rate']=data1['travel_time_40_59_rate'].fillna((data1['travel_time_40_59_rate'].mean()))
data1['travel_time_60_89_rate']=data1['travel_time_60_89_rate'].fillna((data1['travel_time_60_89_rate'].mean()))
data1['health_insurance_public_rate']=data1['health_insurance_public_rate'].fillna((data1['health_insurance_public_rate'].mean()))
data1['health_insurance_none_rate']=data1['health_insurance_none_rate'].fillna((data1['health_insurance_none_rate'].mean()))
data1['age_median']=data1['age_median'].fillna((data1['age_median'].mean()))
data1['travel_time_average']=data1['travel_time_average'].fillna((data1['travel_time_average'].mean()))
data1['voting_2012_dem_percentage']=data1['voting_2012_dem_percentage'].fillna((data1['voting_2012_dem_percentage'].mean()))
data1['voting_2012_gop_percentage']=data1['voting_2012_gop_percentage'].fillna((data1['voting_2012_gop_percentage'].mean()))
data1['number_of_years_of_education']=data1['number_of_years_of_education'].fillna((data1['number_of_years_of_education'].mean()))
data1['diversity']=data1['diversity'].fillna((data1['diversity'].mean()))
data1['number_of_solar_system_per_household']=data1['number_of_solar_system_per_household'].fillna((data1['number_of_solar_system_per_household'].mean()))
np.sum(data1.isnull())
```

```python
from scipy import stats
deepsolar=data1.drop(['county', 'voting_2016_dem_win', 'voting_2012_dem_win', 'state'], 1)
z = np.abs(stats.zscore(deepsolar._get_numeric_data()))
print(z)
data1=data1[(z<3).all(axis=1)]
print(data1.shape)
```
